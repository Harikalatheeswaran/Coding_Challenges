### 1. Longest Palindromic Sub-String : 

- [Link](https://leetcode.com/problems/longest-palindromic-substring/)
- `Time Taken` - `4644`s

```python
def longestPalindrome(self, s: str) -> str:
    # defining the funtion to check for palindrome
    def is_palin(s:str) -> bool:
        if s[::-1] == s:
            return True
        else : return False

    n = len(s)
    # if s is single charcter, return that
    if n == 1: return s

    # if s is a palindrome, return that
    if is_palin(s): return s

    palindromes = []
    for i in range(2, n):
        for j in range(n):
            if j+i <= n:
                if is_palin(s[j:j+i]) : 
                    palindromes.append(s[j:j+i])
                    
            else : break
    
    palindromes.sort(key=len)

    if len(palindromes) >= 1:
        return palindromes[-1]
    else : return s[0]
            
        
```
